This study sought to investigate the relationship between market volatility, growth metrics, and ESG performance. To operationalize this, two distinct hypothetical portfolios were constructed based on S&P 500 data: a "Defensive portfolio" (comprising the lowest volatility decile) and an "Aggressive portfolio" (comprising the highest 1 year return decile). The hypothesis tested was whether financial stability acts as a proxy for long term sustainability planning.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

df = pd.read_csv("/content/esg_financial_dataset_with_sector.csv")
df_clean = df.dropna(subset=["ESG_Rating_Score", "P/E Ratio"])
df_clean = df_clean[(df_clean["P/E Ratio"] < 200) & (df_clean["P/E Ratio"] > 0)]

low_vol = df_clean.sort_values("Avg Volatility (30D)").head(50)
high_growth = df_clean.sort_values("1Y Return",ascending=False).head(50)
hg_pe_mean = high_growth["P/E Ratio"].mean()
lv_pe_mean = low_vol["P/E Ratio"].mean()
hg_esg_score_mean = high_growth["ESG_Rating_Score"].mean()
lv_esg_score_mean = low_vol["ESG_Rating_Score"].mean()

portfolio_summary = pd.DataFrame({
    "Portfolio": ["Low Volatility (Defensive)", "High Growth (Aggressive)"],
    "Mean ESG Score": [lv_esg_score_mean, hg_esg_score_mean],
    "Mean P/E Ratio": [lv_pe_mean, hg_pe_mean]
})

print("Confronto Portafogli:")
print(portfolio_summary)


plt.figure(figsize=(10, 6))
sns.barplot(data=portfolio_summary, x="Portfolio", y="Mean ESG Score", palette=["#2ecc71", "#e74c3c"])
overall_mean = df_clean["ESG_Rating_Score"].mean()
plt.axhline(y=overall_mean, color='blue', linestyle='--', label=f"S&P 500 Average ({overall_mean:.1f})")
plt.title("ESG Score Comparison: Defensive vs Aggressive")
plt.ylabel("Average ESG Score (0-10)")
plt.legend()
plt.show()
```

The comparative analysis highlights a strong synergy between low volatility and high ESG performance.
The Defensive portfolio yielded a significantly higher average ESG score (7.86) compared to the Aggressive portfolio (6.80). Furthermore, from a valuation perspective, the defensive portfolio trades at a more attractive P/E ratio (26x vs. 45x). These results suggest that the "Low Volatility" factor naturally overlaps with the "ESG" factor.
Therefore, investors prioritizing capital preservation do not need to pay a "green premium"; rather, they acquire a more sustainable portfolio, reinforcing the view of ESG as a risk mitigation tool rather than merely an ethical overlay.
